# Permutações de *n* objetos

Notebook Aula 09-01.png

```python
from itertools import permutations, combinations
from random import choices

# Chave para permutação
chave = 'Python'

# Gerção de todas as permutações possíveis
perm_chave = permutations(chave)

# Exibição de 20 permutações sorteadas aleatoriamente
#list = lista todas as permutações
lista_perm = list(perm_chave)
print(f'São possíveis {len(lista_perm)} permutações para os caracteres de "{chave}".')
print('Seguem 20 permutações sorteadas aleatoriamente:')
#k = quantos sorteios fazer
sorteio = choices(lista_perm, k=20)
for caso in sorteio:
  print(caso)
```

# Permutações distintas de *n* objetos
Quando há repetição nos elementos da chave de permutação (que sempre deve ser um iterável), haverá também permutações iguais. Se desejarmos apenas as permutações distintas, é preciso usar a função `distinct_permutations` do módulo `more_itertools`.

```python
from more_itertools import distinct_permutations

# Chave para permutação
chave = 'Geek'

# Geração de todas as permutações possíveis
todas_perm = list(permutations(chave))

# Exibição de todas as permutações
print(f'Existem {len(todas_perm)} permutações possíveis para "Geek".')
for i in range(6):
  print(todas_perm[i*4:i*4+4])
```

```python
# Geração das permutações distintas
perm_dist = list(distinct_permutations(chave))

# Exibição das permutações distintas
print(f'Existem {len(perm_dist)} permutações distintas possíveis para "Geek".')
for i in range(3):
  print(perm_dist[i*4:i*4+4])
```

# Arranjo de *n* objetos tomados *r* a *r*
Notebook Aula 09-02.png

```python
# Chave para permutação
chave = ['BRA', 'CHI', 'CAN', 'USA', 'ARG']


#arranjo considera a ordem dos elementos. ex: (BRA, CHI, USA) e (CHI, BRA, USA) são elementos diferentes
# Geração dos arranjos dos 5 objetos de chave, tomados 3 a 3
perm_chave = list(permutations(chave, 3))

# Exibição dos arranjos, tomados 3 a 3
print(f'Há {len(perm_chave)} elementos nos arranjos de {chave}, tomados 3 a 3.')
for i in range(15):
  print(perm_chave[i*4:i*4+4])
```

# Combinações de *n* objetos, tomados *r* a *r*
Se a ordem de cada permutação não for importante (ABC é o mesmo que BAC, por exemplo), o número de maneiras de selecionar *r* dos *n* objetos sem levar em consideração a ordem é chamada de combinação de *n* objetos tomados *r* a *r*.

Notebook Aula 09-03.png

```python
# Geração das combinações dos 5 objetos de chave, tomados 3 a 3
comb_chave = (list(combinations(chave, 3)))

#combinação não considera a ordem dos elementos. exemplo: (BRA, CHI, CAN) e (CAN, USA, BRA) são iguais;
# Exibição das combinações, tomadas 3 a 3
print(f'Há {len(comb_chave)} elementos nas combinações de {chave}, tomadas 3 a 3.')
for i in range(5):
  print(comb_chave[i*2:i*2+2])
```
